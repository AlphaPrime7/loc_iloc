```python
#LOC
import pandas as pd
import numpy as np

#frame 1
df = pd.DataFrame(
    np.zeros((10,3)),
    columns=['up','down','right']
)

#base indexing - similar to R
df.loc[1:3]
#conditional indexing
df.loc[(df.up == 0) & (df.down == 0)]
#update data
df.loc[(df.up == 0) , ['up']] = 2
display(df) #super enter = ctrl + enter

#frame 2
df1 = pd.DataFrame(
    np.zeros((10,3)),
    columns=['up','down','right']
)

#merge frames 
#merge
jkey = list(df.columns)
dfm = pd.merge(df,df1, how='left', on=jkey)
display(dfm)
#concat
frames = [df,df1]
dfm = pd.concat(frames)
display(dfm)
dfm.tail()
```

```python
#ILOC
import pandas as pd
import numpy as np

#frame 1
df = pd.DataFrame(
    np.zeros((10,3)),
    columns=['up','down','right']
)
df.info()

#base indexing
df.iloc[0:len(df.axes[0])] #just len(df) or df.shape[0] or len(df.index) or df[df.columns[0]].count(); 0 = rows, 1 = columns
#discrete indexing
df.iloc[[0, 1, 2]]
#discrete indexing with the :
df.iloc[1, :]
```

```python
n=6
def rng(n):
    s = []
    for i in range(6):
        s.append(i)
    return s
rng(n)
```

```python
#GUARD RAILS (TOP VS BOTTOM)
import pandas as pd
import numpy as np

#frame 1
df = pd.DataFrame(
    np.zeros((10,3)),
    columns=['up','down','right']
)
print(len(df.columns))
print(rng(len(df.axes[1])))

#iloc
df.iloc[-1] # or display(df.index[-1]) or display(df[-1:]); last row
df.iloc[0] #or print(df.head(1)) or df.iloc[:1] or df['up'].iloc[:1] or print(df.values[:1]) or df.at[0, 'up']; first rows
#iat(index at)-useful where -1 and 0 can be replace with iterands
df.iat[0,0]
df.iat[-1,0]
```
